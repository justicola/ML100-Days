```python
#先看一下資料概況
app_train = pd.DataFrame(app_train)
print(app_train.shape)
app_train.head()
```

```python
#取單一欄位 dfName['xxx'](series)，取2以上欄位dfName[['xxx','ooo'...]](dataframe)
sub_train = pd.DataFrame(app_train['WEEKDAY_APPR_PROCESS_START'])
print(sub_train.shape)
sub_train.head()
```

```python
# LabelEncoder()
# 加上新欄位'xxxx'(名字自己命)，這個欄位資料來自於...(對應的欄位名稱)
# :表示把所有的 row抓出來，1:3則表示抓 index 1~2的 row

sub_train['rWEEKDAY_APPR_PROCESS_START'] = le.fit_transform(sub_train.loc[:,'WEEKDAY_APPR_PROCESS_START'])
sub_train.head(20)
```

```python
#建個範例來試試
import pandas as pd
dict_brics = {
    'country':["Brazil", "Russia", "India", "China", "South Africa"],
    'capital':["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
    "area":[8.516, 17.10, 3.286, 9.597, 1.221],
    "population":[200.4, 143.5, 1252, 1357, 52.98]}

# brics = pd.DataFrame(dict_brics)
brics = pd.DataFrame(dict_brics,index = ['BR','RU','IN','CH','SA'])
#如果沒有設定index，預設就是0,1,2,3...
print(brics,'\n')


# 有 index_label可以直接 Q資料(Q:資料很多筆的時候怎麼自建 index?? 
# 用 index_label選資料是選一個 row，用.loc選，一組 []是 series type
# loc 是 label-based 的取資料方式：需要使用 index_label來取得 row 的資料。
# iloc 則是 index-based 的取資料方式：使用數字的 index來取得 row 的資料，這點和其他語言比較像。

print(brics.loc['CH'])
print('='*50)

# 兩組 [[]] 把資料從 series轉成 df type
print(brics.loc[['CH']])
```
